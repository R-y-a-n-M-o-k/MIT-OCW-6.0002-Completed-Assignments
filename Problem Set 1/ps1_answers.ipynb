## 2. Greedy and Brute Force algorithms

```python
cow_list = cow.load_cows('ps1_cow_data.txt')
print(cow_list)
print()
greedy_trips = cow.greedy_cow_transport(cow_list,10)
for i in greedy_trips:
    print(i)
    print()
```

```python
global_trip = cow.brute_force_cow_transport(cow_list,limit=10)
for i in global_trip:
    print(i)
    print()
```

There is a difference in the number of trips in the greedy and brute force algorithms (brute force is lower) for ps1_cow_data. For large cows such as 'Betsy' and 'Henrietta', both yield the same trip - unsurprising since the only way they can be loaded is by themselves.

Coincidentally both 'Maggie' and 'Herman' are placed on the same trip in both solutions. The difference then comes in cows with low to mid size. There is a flaw in the greedy algorithm, where a cow or multiple cows with medium size (but not large enough to rule out any other cows) prevents the smaller cows from filling the gaps and the trip becomes suboptimal globally. 

```python
cow_list = cow.load_cows('ps1_cow_data_2.txt')
print(cow_list)
greedy_trips = cow.greedy_cow_transport(cow_list,limit=10)
for i in greedy_trips:
    print(i)
    print()
```

```python
global_trip = cow.brute_force_cow_transport(cow_list,limit=10)
for i in global_trip:
    print(i)
    print()
```

Here both the global and greedy algorithm achieve the same trip length, although the solutions are not the same. The luckiness in the degeneracy (of algorithms) arises from the fact the greedy algorithm just so happens to be able to clear its midsized cows ('Dottie' and 'Milkshake') which allows the small cows to optimally pack together. Greedy algorithm acheives a globally optimal, but not uniquely optimal solution.

**1. Explain why it would be difficult to use a brute force algorithm to solve this problem if there
   were 30 different egg weights. You do not need to implement a brute force algorithm in order to
   answer this**
   
- The general brute force approach in solving is to recursively take the target weight and subtract by each of the egg weights. The minimum number of eggs is then 1 + the smallest solution out of the subtracted weights (number of eggs required is smallest). 
- As we do this recursively, for each level we go down, we have to call the function again for each egg weight. Therefore the number of calls grows exponentially with recursion level. The size of the computation is large if the target weight is much larger than the eggs weights as we end up with many levels down. 
- For 30 egg weights, the time complexity grows roughly as 30^n where n is the level of recursion required to end up at the trivial weights of 0 and 1.
